```python
##################################  PREPROCESSING OF DATA  ###############################

print(df.isnull().sum())
#print(df.loc[df.GENDER.isnull()])
#print(p.isnull(df.GENDER))

########################### DROP COLUMNS  ################################################

df=df.drop(['SYMBOL','GENERAL\nVOTES','POSTAL\nVOTES'],axis=1)
df=df.drop(['NAME'],axis=1)

df.shape
```

```python
######################### CONVERTING CATEGORICAL VALUES INTO NUMERICAL AND FILLING NULL VALUES #####################


df.GENDER.loc[df.GENDER=='MALE']=0
df.GENDER.loc[df.GENDER=='FEMALE']=1
df['GENDER']=df['GENDER'].fillna(df['GENDER'].median())
print(df.GENDER.value_counts())

```

```python
df['AGE']=df['AGE'].fillna(df['AGE'].median())

print(df.CATEGORY.value_counts())
print('\n',df.EDUCATION.value_counts(),'\n')

print(df['CRIMINAL\nCASES'].value_counts())
df=df.drop(df.loc[df['CRIMINAL\nCASES']=='Not Available'].index)
df['CRIMINAL\nCASES']=df['CRIMINAL\nCASES'].fillna(df['CRIMINAL\nCASES'].median())

print(df.isnull().sum())
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoderx=LabelEncoder()

df.PARTY=encoderx.fit_transform(df.PARTY)
print(df.PARTY.value_counts())
print(df.isnull().sum())


print(df.CATEGORY.value_counts())
df['CATEGORY']=df['CATEGORY'].fillna(df.CATEGORY.mode()[0])
#df['CATEGORY']=df['CATEGORY'].replace(to_replace=0,value='nl')
print('\n MODE  - ',df.CATEGORY.mode()[0])


encodery=LabelEncoder()
df.CATEGORY=encodery.fit_transform(df.CATEGORY)
df.isnull().sum()
print(df.CATEGORY.value_counts())
```

```python
print(df.EDUCATION.value_counts())
print(df.isnull().sum())


df.EDUCATION.loc[df.EDUCATION.isin(['Post Graduate\n'])]='Post Graduate'
arr=['8th Pass','Illiterate','Literate','5th Pass']
df.EDUCATION.loc[df.EDUCATION.isin(arr)]='Below 8th'
```

```python
df.EDUCATION=df.EDUCATION.fillna('Others')

encoderx=LabelEncoder()
df.EDUCATION=encoderx.fit_transform(df.EDUCATION)
```

```python
df=df.drop(['ASSETS','LIABILITIES'],axis=1)

encoderl=LabelEncoder()
df.STATE=encoderl.fit_transform(df.STATE)
df.CONSTITUENCY=encoderl.fit_transform(df.CONSTITUENCY)
print(df.CONSTITUENCY.value_counts())
```

```python
################################ SPLIT DATA INTO TRAIN AND TEST DATA ######################################


x=df.drop('WINNER',axis=1)
y=df.WINNER

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
print(xtest)
print(xtrain)


```

```python
################################## TRAINING THE MODEL ################################################### 


from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics


LRmodel = LogisticRegression()
DTmodel = DecisionTreeClassifier(random_state = 114 )
RFmodel = RandomForestClassifier(n_estimators = 50 , criterion = 'entropy' , random_state = 114 )
KNNmodel = KNeighborsClassifier(n_neighbors = 3 )
SVMmodel = SVC(kernel = 'rbf', random_state = 114)


LRresult = LRmodel.fit( xtrain , ytrain )
predictions=LRresult.predict(xtest)
print('\n\n LR ACCURACY = ',metrics.accuracy_score(ytest,predictions))

DTresult = DTmodel.fit( xtrain , ytrain )
predictions=DTresult.predict(xtest)
print('\n\n DT ACCURACY = ',metrics.accuracy_score(ytest,predictions))

RFresult = RFmodel.fit( xtrain , ytrain )
predictions=RFresult.predict(xtest)
print('\n\n RF ACCURACY = ',metrics.accuracy_score(ytest,predictions))

KNNresult = KNNmodel.fit( xtrain , ytrain )
predictions=KNNresult.predict(xtest)
print('\n\n KNN ACCURACY = ',metrics.accuracy_score(ytest,predictions))

SVMresult = SVMmodel.fit( xtrain , ytrain )
predictions=SVMresult.predict(xtest)
print('\n\n SVM ACCURACY = ',metrics.accuracy_score(ytest,predictions))



```

```python
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(ytest, predictions)
print('CONFUSION MATRIX   - \n\n ',cm,'\n\n')
print ( ' \n **Classification Report: \n ' ,metrics . classification_report(ytest,predictions))

sn.heatmap(cm,annot=True,fmt='2.0f')



### CONCLUSION - RANDOM FOREST IS THE BEST SUITED MODEL WITH ACCURACY 94.8%.
```
